Triangle	 	$T_n$ = n(n+1)/2	 	1, 3, 6, 10, 15, 21, 28, 36, 45, ...<br>
Pentagonal	 	$P_n$ = n(3n−1)/2	 	1, 5, 12, 22, 35, ...<br>
Hexagonal	 	$H_n$ = n(2n−1)	 	1, 6, 15, 28, 45, ...<br>
